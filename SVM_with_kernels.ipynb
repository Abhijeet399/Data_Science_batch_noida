```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Generating our data

X, Y = make_circles(n_samples = 500, noise = 0.02)

plt.scatter(X[:,0], X[:, 1], c=Y, marker = '.')
plt.show()
```

```python
print(X.shape)
```

```python
# ADD new dimension to X
X1 = X[:,0].reshape((-1,1))
X2 = X[:,1].reshape((-1,1))
X3 = (X1**2 + X2**2)

X = np.hstack((X, X3))

# Visualization
fig = plt.figure(figsize = (10,10))
axes = fig.add_subplot(111, projection = '3d')
axes.scatter(X1, X2, X3, c=Y, depthshade = True)
plt.show()
```

```python
from sklearn import svm

svc = svm.SVC(kernel = 'linear')
svc.fit(X, Y)
print(svc.score(X, Y))

w = svc.coef_
b = svc.intercept_
```

```python
#Plotting hyperplane

x1 = X[:,0].reshape((-1,1))
x2 = X[:,1].reshape((-1,1))
x1, x2 = np.meshgrid(x1, x2)
x3 = -(w[0][0]*x1 + w[0][1]*x2 + b)/w[0][2]

fig = plt.figure(figsize = (10,10))
axes2 = fig.add_subplot(111, projection = '3d')
axes2.scatter(X1, X2, X3, c=Y, depthshade = True)
axes1 = fig.add_subplot(111, projection = '3d')
axes1.scatter(x1, x2, x3, alpha = 0.5)
plt.show()
```

```python
from sklearn import svm

svc = svm.SVC(kernel = 'poly')
svc.fit(X, Y)
print(svc.score(X, Y))
```

```python
from sklearn import svm

svc = svm.SVC(kernel = 'rbf')
svc.fit(X, Y)
print(svc.score(X, Y))
```

```python
def custom_kernel(x, z):
  return np.square(np.dot(x, z.T))
```

```python
svc = svm.SVC(kernel = custom_kernel)
svc.fit(X, Y)
print(svc.score(X, Y))
```
