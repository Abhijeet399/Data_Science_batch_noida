```python
class fib:
    def __init__(self):
        self.prev = 0
        self.curr = 1
        
    def __iter__(self):
        return self
    
    def __next__(self):
        value = self.curr
        self.curr += self.prev
        self.prev = value
        return value
```

```python
f = fib()
```

```python
f_iter = iter(f)
```

```python
next(f_iter)
```

```python
def fibo():
    prev, curr = 0, 1
    while True:
        yield curr
        prev, curr = curr, prev+curr
```

```python
type(fibo())
```
