```python
plt.style.use('ggplot')
plt.figure(figsize = (5, 5))
plt.scatter(x, y1, color = 'red', label = 'X^2', marker = '^', linestyle='--')
plt.xlabel('x')
plt.ylabel('Y')
plt.title('tittle')
plt.legend()
plt.show()
```

```python
x_cor = np.array([0, 1, 2])
# x_cor_another = [1,2,3]
plt.bar(x_cor, [20, 10, 45], width=0.2, color='blue', label='H', tick_label=["x", "y", "z"])
plt.bar(x_cor + 0.5, [35, 3, 15], width=0.2, color='red', label='G', tick_label=["a", "b", "c"])
plt.xlim(-0.5, 3)
plt.ylim(0, 50)
plt.show()
```
